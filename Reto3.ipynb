Eliminar duplicados

```python
print("Duplicados Antes:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicados Después:", df.duplicated().sum())
```

Eliminar nulos

```python
df = df.dropna()
df.isnull().sum()
df.info()
```
